```python
import pandas as pd
import matplotlib.pyplot as plt

# load dataset
df = pd.read_csv("/content/sample_data/preprocessing_practice_dataset.csv")

# count employees in each department
dept_count = df['Department'].value_counts()

# pie chart
plt.figure(figsize=(8,8))
plt.pie(
    dept_count,
    labels=dept_count.index,
    autopct='%1.1f%%'
)

plt.title('Department Distribution')
plt.show()
```

```python
from sklearn.preprocessing import MinMaxScaler,StandardScaler
file_path='/content/sample_data/preprocessing_practice_dataset.csv'
prepos_data=pd.read_csv(file_path)
numerical_columns=['Employee_ID','Age','Experience_Years','Monthly_Salary']
prepos_data.dropna(subset=numerical_columns,inplace=True)
minmax_scaler=MinMaxScaler()
standard_scaler=StandardScaler()
prepos_data['Employee_ID_MinMax']=minmax_scaler.fit_transform(prepos_data[['Employee_ID']])
prepos_data['Age_MinMax']=minmax_scaler.fit_transform(prepos_data[['Age']])
prepos_data['Experience_Years_MinMax']=minmax_scaler.fit_transform(prepos_data[['Experience_Years']])
prepos_data['Monthly_Salary_MinMax']=minmax_scaler.fit_transform(prepos_data[['Monthly_Salary']])

prepos_data['Employee_ID_Standard']=standard_scaler.fit_transform(prepos_data[['Employee_ID']])
prepos_data['Age_Standard']=standard_scaler.fit_transform(prepos_data[['Age']])
prepos_data['Experience_Years_Standard']=standard_scaler.fit_transform(prepos_data[['Experience_Years']])
prepos_data['Monthly_Salary_Standard']=standard_scaler.fit_transform(prepos_data[['Monthly_Salary']])
```
