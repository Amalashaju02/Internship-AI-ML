```python
df['Age_mean_imputed'] = df['Age'].fillna(df['Age'].mean())
df['Fare_mean_imputed'] = df['Fare'].fillna(df['Fare'].mean())
df['Cabin_imputed'] = df['Cabin'].fillna('Unknown')

display(df[['Age', 'Age_mean_imputed', 'Fare', 'Fare_mean_imputed', 'Cabin', 'Cabin_imputed']].head())
```

```python
display(df[['Age_mean_imputed', 'Fare_mean_imputed', 'Cabin_imputed']].isnull().sum())
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# load the dataset
df = pd.read_csv("/content/sample_data/titanic uncleaned.csv")

# set up the plot style
sns.set(style="whitegrid")

plt.figure(figsize=(10, 6))
sns.boxplot(data=df,x='Pclass',y='Age',palette='pastel')

# set plot title and labels
plt.title('Boxplot of Age by Pclass', fontsize=16)
plt.xlabel('Pclass', fontsize=12)
plt.ylabel('Age', fontsize=12)

# show the plot
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# load the dataset
df = pd.read_csv("/content/sample_data/titanic uncleaned.csv")

# set up the plot style
sns.set(style="whitegrid")

plt.figure(figsize=(10, 6))
sns.barplot(data=df,x='Pclass',y='Age',palette='pastel')

# set plot title and labels
plt.title('Barplot of Age by Pclass', fontsize=16)
plt.xlabel('Pclass', fontsize=12)
plt.ylabel('Age', fontsize=12)

# show the plot
plt.show()
```
